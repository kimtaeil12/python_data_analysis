```python
csData = pd.read_csv("../dataset/kopo_customerdata.csv")
```

```python
csTotallList = csData["TOTAL_AMOUNT"].tolist()
```

```python
maxValue = max(csTotallList)
minValue = min(csTotallList)
print(maxValue, minValue)
```

```python
minValueCnt = csTotallList.count(minValue)
maxValueCnt = csTotallList.count(maxValue)
print(minValueCnt, maxValueCnt)
```

```python
denominator  = sum(csTotallList) - (minValue*minValueCnt) - (maxValue*maxValueCnt)
numerator  = len(csTotallList) - minValueCnt - maxValueCnt
```

```python
answer = int(denominator / numerator)
answer
```
